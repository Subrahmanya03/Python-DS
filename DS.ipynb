```python
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items.append(item)

    def pop(self):
        if not self.is_empty():
            return self.items.pop()
        return None

    def peek(self):
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)


stack = Stack()
stack.push(1)
stack.push(2)
stack.push(3)

print("Stack Size:", stack.size())
print("Top Element:", stack.peek())
print("Popped Element:", stack.pop())
print("Stack Size After Pop:", stack.size())
```

```python
import random
import time
import matplotlib.pyplot as plt

# Dictionary to store function execution times
timings = {
    'append': [],
    'reverse': [],
    'shuffle': [],
    'delete_beginning': [],
    'delete_end': [],
    'delete_middle': [],
    'delete_at': []
}

# Define a decorator to time functions
def timeit(method):
    def timed(*args, **kwargs):
        start_time = time.time()
        result = method(*args, **kwargs)
        end_time = time.time()
        execution_time = end_time - start_time
        method_name = method.__name__
        if method_name in timings:
            timings[method_name].append(execution_time)
        return result
    return timed

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    @timeit
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    @timeit
    def reverse(self):
        prev = None
        current = self.head
        while current:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        self.head = prev

    @timeit
    def shuffle(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(current)
            current = current.next
        
        random.shuffle(nodes)
        
        if not nodes:
            self.head = None
            return
        
        self.head = nodes[0]
        current = self.head
        for node in nodes[1:]:
            current.next = node
            current = node
        current.next = None

    @timeit
    def delete_beginning(self):
        if self.head is None:
            return
        self.head = self.head.next

    @timeit
    def delete_end(self):
        if self.head is None:
            return
        if self.head.next is None:
            self.head = None
            return
        current = self.head
        while current.next and current.next.next:
            current = current.next
        current.next = None

    @timeit
    def delete_middle(self):
        if self.head is None or self.head.next is None:
            return
        slow = self.head
        fast = self.head
        prev = None
        while fast and fast.next:
            fast = fast.next.next
            prev = slow
            slow = slow.next
        if prev:
            prev.next = slow.next

    @timeit
    def delete_at(self, location):
        if self.head is None:
            return
        if location == 0:
            self.delete_beginning()
            return
        current = self.head
        prev = None
        count = 0
        while current and count < location:
            prev = current
            current = current.next
            count += 1
        if current:
            prev.next = current.next

    def display(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Function to run and measure timings
def run_linked_list_operations():
    ll = LinkedList()
    # Append multiple items
    for i in range(100):  # Adjust the range for more data
        ll.append(i)

    ll.reverse()
    ll.shuffle()
    ll.delete_beginning()
    ll.delete_end()
    ll.delete_middle()
    ll.delete_at(0)

if __name__ == "__main__":
    run_linked_list_operations()

    # Compute average timings for each operation
    avg_timings = {method: sum(times)/len(times) if times else 0 for method, times in timings.items()}
    for method_name, avg_time in avg_timings.items():
        print(f"{method_name}: {avg_time:.6f} seconds")

    # Plot timings
    methods = list(avg_timings.keys())
    times = list(avg_timings.values())

    plt.figure(figsize=(12, 8))
    plt.barh(methods, times, color='skyblue')
    plt.xlabel('Time (seconds)')
    plt.title('Average Execution Time of Linked List Operations')
    plt.gca().invert_yaxis()
    plt.show()
```
